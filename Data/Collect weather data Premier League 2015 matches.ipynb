```python
import requests
import os
import pandas as pd
import numpy as np
```

```python
path_val = 'Weather/match_weather_data/'
```

```python
directory='Football/all_match_details.csv'
df=pd.read_csv(directory,sep=',')
df.drop('Unnamed: 0', axis=1, inplace=True)
len(df)
```

```python
df['FeelsLikeF']=np.nan
df['HeatIndexF']=np.nan
df['humidity']=np.nan
df['visibility']=np.nan
df['weatherDesc']=np.nan
df['windspeedMiles']=np.nan
df.head()
```

```python
#weatherResponse
```

```python
import json
import time
key =  os.getenv("key")
#key='e62ae1c5bb104384ae973054171604'
host='http://api.worldweatheronline.com/premium/v1/past-weather.ashx?'
match_date_list = df[['Date','Location','home','away']].values.tolist()
for i in match_date_list:
    city= i[1]
    start_date=i[0]
    home_team=i[2]
    away_team=i[3]
    inputRequestString=host+"key="+key+"&q="+city+",united+kingdom"+"&date="+start_date+"&format=json&tp=24"
    weatherResponse=requests.get(inputRequestString).json()
    file_name =path_val+city+'/'+city+'_'+start_date+'.json'
    os.makedirs(os.path.dirname(file_name), exist_ok=True)
    with open(file_name, 'w', encoding='utf-8') as f:
        json.dump(weatherResponse, f)
        f.close()
    
    FeelsLikeF=weatherResponse['data']['weather'][0]['hourly'][0]['FeelsLikeF']
    HeatIndexF=weatherResponse['data']['weather'][0]['hourly'][0]['HeatIndexF']
    humidity=weatherResponse['data']['weather'][0]['hourly'][0]['humidity']
    visibility=weatherResponse['data']['weather'][0]['hourly'][0]['visibility']
    weatherDesc=weatherResponse['data']['weather'][0]['hourly'][0]['weatherDesc'][0]['value']
    windspeedMiles=weatherResponse['data']['weather'][0]['hourly'][0]['windspeedMiles']
    
    df.loc[(df['Date'] == start_date) &(df['Location']== city)& (df['home']== home_team), 'FeelsLikeF'] = FeelsLikeF
    df.loc[(df['Date'] == start_date) & (df['Location']== city)& (df['home']== home_team), 'HeatIndexF'] = HeatIndexF
    df.loc[(df['Date'] == start_date) & (df['Location']== city)& (df['home']== home_team), 'humidity'] = humidity
    df.loc[(df['Date'] == start_date) & (df['Location']== city)& (df['home']== home_team), 'visibility'] = visibility
    df.loc[(df['Date'] == start_date) & (df['Location']== city)& (df['home']== home_team), 'weatherDesc'] = weatherDesc
    df.loc[(df['Date'] == start_date) & (df['Location']== city)& (df['home']== home_team), 'windspeedMiles'] = windspeedMiles
    
    
#weatherResponse['data']['weather'][0]['hourly'][0]['weatherDesc'][0]['value']#['hourly']
#"http://history.openweathermap.org/data/2.5/history/city?q="+city+",UK"+"&type=hour&start="+start_date
#competition_list
```
