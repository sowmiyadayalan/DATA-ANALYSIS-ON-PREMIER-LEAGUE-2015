```python
directory="Data/Football/Premier_League_2015_Data"
count=0
match_dic = {}
match_details_dataframe = pd.DataFrame(columns = ['team','match_count','win','draw','loss','for_goals','against_goal'])
home_match_dataframe = pd.DataFrame(columns = ['team','match_count','win','draw','loss','for_goals','against_goal'])
away_match_dataframe = pd.DataFrame(columns = ['team','match_count','win','draw','loss','for_goals','against_goal'])
for root, dirs, files in os.walk(directory):
     for file in files:
        with open(os.path.join(root, file), "r") as f:
            data = json.load(f)
            if (match_details_dataframe['team']== data['homeTeamName']).any():
                #print("check")
                match_details_dataframe.loc[match_details_dataframe['team'] == data['homeTeamName'], 'match_count'] += 1
                match_details_dataframe.loc[match_details_dataframe['team'] == data['homeTeamName'], 'for_goals'] += data['result']['goalsHomeTeam']
                match_details_dataframe.loc[match_details_dataframe['team'] == data['homeTeamName'], 'against_goal'] += data['result']['goalsAwayTeam']
                if data['result']['goalsHomeTeam']> data['result']['goalsAwayTeam']:
                    match_details_dataframe.loc[match_details_dataframe['team'] == data['homeTeamName'], 'win'] += 1
                elif data['result']['goalsHomeTeam']< data['result']['goalsAwayTeam']:
                    match_details_dataframe.loc[match_details_dataframe['team'] == data['homeTeamName'], 'loss'] += 1
                else:
                    match_details_dataframe.loc[match_details_dataframe['team'] == data['homeTeamName'], 'draw'] += 1
                 
            else:
                count += 1
                win = 0
                loss = 0
                draw = 0
                match_count = 1
                for_goals=data['result']['goalsHomeTeam']
                against_goal=data['result']['goalsAwayTeam']
                if data['result']['goalsHomeTeam']> data['result']['goalsAwayTeam']:
                    win += 1
                elif data['result']['goalsHomeTeam']< data['result']['goalsAwayTeam']:
                    loss += 1
                else:
                    draw += 1
                match_details_dataframe.loc[count] = [data['homeTeamName'],match_count,win,draw,loss,for_goals,against_goal]
            ######################
            if (match_details_dataframe['team']== data['awayTeamName']).any():
                #print("check")
                match_details_dataframe.loc[match_details_dataframe['team'] == data['awayTeamName'], 'match_count'] += 1
                match_details_dataframe.loc[match_details_dataframe['team'] == data['awayTeamName'], 'for_goals'] += data['result']['goalsAwayTeam']
                match_details_dataframe.loc[match_details_dataframe['team'] == data['awayTeamName'], 'against_goal'] +=data['result']['goalsHomeTeam'] 
                if data['result']['goalsHomeTeam']< data['result']['goalsAwayTeam']:
                    match_details_dataframe.loc[match_details_dataframe['team'] == data['awayTeamName'], 'win'] += 1
                elif data['result']['goalsHomeTeam']> data['result']['goalsAwayTeam']:
                    match_details_dataframe.loc[match_details_dataframe['team'] == data['awayTeamName'], 'loss'] += 1
                else:
                    match_details_dataframe.loc[match_details_dataframe['team'] == data['awayTeamName'], 'draw'] += 1
                 
            else:
                count += 1
                win = 0
                loss = 0
                draw = 0
                match_count = 1
                for_goals=data['result']['goalsAwayTeam']
                against_goal=data['result']['goalsHomeTeam']
                if data['result']['goalsHomeTeam']< data['result']['goalsAwayTeam']:
                    win += 1
                elif data['result']['goalsHomeTeam']> data['result']['goalsAwayTeam']:
                    loss += 1
                else:
                    draw += 1
                match_details_dataframe.loc[count] = [data['awayTeamName'],match_count,win,draw,loss,for_goals,against_goal]
# Home Match league table            
            if (home_match_dataframe['team']== data['homeTeamName']).any():
                #print("check")
                home_match_dataframe.loc[home_match_dataframe['team'] == data['homeTeamName'], 'match_count'] += 1
                home_match_dataframe.loc[home_match_dataframe['team'] == data['homeTeamName'], 'for_goals'] += data['result']['goalsHomeTeam']
                home_match_dataframe.loc[home_match_dataframe['team'] == data['homeTeamName'], 'against_goal'] += data['result']['goalsAwayTeam']
                if data['result']['goalsHomeTeam']> data['result']['goalsAwayTeam']:
                    home_match_dataframe.loc[home_match_dataframe['team'] == data['homeTeamName'], 'win'] += 1
                elif data['result']['goalsHomeTeam']< data['result']['goalsAwayTeam']:
                    home_match_dataframe.loc[home_match_dataframe['team'] == data['homeTeamName'], 'loss'] += 1
                else:
                    home_match_dataframe.loc[home_match_dataframe['team'] == data['homeTeamName'], 'draw'] += 1
                 
            else:
                count += 1
                win = 0
                loss = 0
                draw = 0
                match_count = 1
                for_goals=data['result']['goalsHomeTeam']
                against_goal=data['result']['goalsAwayTeam']
                if data['result']['goalsHomeTeam']> data['result']['goalsAwayTeam']:
                    win += 1
                elif data['result']['goalsHomeTeam']< data['result']['goalsAwayTeam']:
                    loss += 1
                else:
                    draw += 1
                home_match_dataframe.loc[count] = [data['homeTeamName'],match_count,win,draw,loss,for_goals,against_goal]
            
# Away Match league table            
            if (away_match_dataframe['team']== data['awayTeamName']).any():
                #print("check")
                away_match_dataframe.loc[away_match_dataframe['team'] == data['awayTeamName'], 'match_count'] += 1
                away_match_dataframe.loc[away_match_dataframe['team'] == data['awayTeamName'], 'for_goals'] += data['result']['goalsAwayTeam']
                away_match_dataframe.loc[away_match_dataframe['team'] == data['awayTeamName'], 'against_goal'] +=data['result']['goalsHomeTeam'] 
                if data['result']['goalsHomeTeam']< data['result']['goalsAwayTeam']:
                    away_match_dataframe.loc[away_match_dataframe['team'] == data['awayTeamName'], 'win'] += 1
                elif data['result']['goalsHomeTeam']> data['result']['goalsAwayTeam']:
                    away_match_dataframe.loc[away_match_dataframe['team'] == data['awayTeamName'], 'loss'] += 1
                else:
                    away_match_dataframe.loc[away_match_dataframe['team'] == data['awayTeamName'], 'draw'] += 1
                 
            else:
                count += 1
                win = 0
                loss = 0
                draw = 0
                match_count = 1
                for_goals=data['result']['goalsAwayTeam']
                against_goal=data['result']['goalsHomeTeam']
                if data['result']['goalsHomeTeam']< data['result']['goalsAwayTeam']:
                    win += 1
                elif data['result']['goalsHomeTeam']> data['result']['goalsAwayTeam']:
                    loss += 1
                else:
                    draw += 1
                away_match_dataframe.loc[count] = [data['awayTeamName'],match_count,win,draw,loss,for_goals,against_goal]
             
            
```

```python
#update consolidated df with goal diff, point and sort them based on points, goal difference and for goals
match_details_dataframe['goal_difference']= match_details_dataframe['for_goals']-match_details_dataframe['against_goal']
match_details_dataframe['points']= (match_details_dataframe['win']*3)+(match_details_dataframe['draw']*1)
match_details_dataframe.sort_values(by=['points','goal_difference','for_goals'],ascending = False).round().reset_index(drop=True).to_csv('Output_Files/overall_league_table.csv', sep=',', encoding='utf-8')
```

```python
#update home match df with goal diff, point and sort them based on points, goal difference and for goals
home_match_dataframe['goal_difference']= home_match_dataframe['for_goals']-home_match_dataframe['against_goal']
home_match_dataframe['points']= (home_match_dataframe['win']*3)+(home_match_dataframe['draw']*1)
home_match_dataframe.sort_values(by=['points','goal_difference','for_goals'],ascending = False).reset_index(drop=True).to_csv('Output_Files/home_matches_league_table.csv', sep=',', encoding='utf-8')
```

```python
#update away matches df with goal diff, point and sort them based on points, goal difference and for goals
away_match_dataframe['goal_difference']= away_match_dataframe['for_goals']-away_match_dataframe['against_goal']
away_match_dataframe['points']= (away_match_dataframe['win']*3)+(away_match_dataframe['draw']*1)
away_match_dataframe.sort_values(by=['points','goal_difference','for_goals'],ascending = False).reset_index(drop=True).to_csv('Output_Files/away_matches_league_table.csv', sep=',', encoding='utf-8')
```
